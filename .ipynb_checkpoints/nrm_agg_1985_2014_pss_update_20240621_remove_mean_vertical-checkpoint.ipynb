## import daily time series from RCMs and GCM
- Directory: \
  `/g/data/tp28/dev/xj2271/acs/eval_1985_2014/supernrm/{model_name}_supernrm_agg_time_series`

```python
variables = ['pr', 'tasmax', 'tasmin']
var_ens   ={
               'pr': 'PRCPTOT',
               'tasmax': 'TX',
               'tasmin':'TN'}

def preprocess(ds):
    ds_new = ds.sel(time=~((ds.time.dt.month == 2) & (ds.time.dt.day == 29)))
    new_time = ds_new.time.dt.strftime("%Y-%m-%d")
    ds_new['time'] = new_time
    return ds_new

    
def process_time_series_nrm(y0, y1, region, sea):
    """
    Parameters:
    -------
        y0: int
            the begin of year [1985]
        y1: int
            the end of year [2014]
        region: string
            'supernrm'
        sea: string
            one season (DJF, JJA)

    Returns
    -------
        agcd: xarray
            daily time series of AGCD for all locations, for each season
        barpa: xarray
            daily time series of BARPA for all locations, all ensembles, and for each season
        ccam: xarray
            daily time series of CCAM for all locations, all ensembles, and for  each season
        icclim: xarray
            daily time series of DES-CCAM for all locations, all ensembles, and for  each season
        cmip6: xarray
            daily time series of CMIP6 for all locations, all ensembles, and for each season
    """
    
    rcm = ['barpa', 'icclim', 'ccam']
    gcm = ['cmip6']
   

    save_dict = '/g/data/tp28/dev/xj2271/acs/' + 'eval_' + str(y0) + '_' + str(y1) + '/' + region + '/'
    barpa = {}
    icclim = {}
    ccam = {}
    cmip6 = {}
    agcd = {}
    rcm = {}
    gcm = {}
    
    filename = 'agg'
    ## read rcm
    for nv, var in enumerate(variables):
        var1 = var_ens[var]
        
        # directory 
        agcd_dir = save_dict + 'agcd/masked_agcd_supernrm_agg_daily_time_series_'+ sea + '_' + var + '*.nc'
       
        barpa_dir = save_dict + 'barpa/barpa_supernrm_agg_daily_time_series_'+ sea + '_' + var + '*.nc'
        
        icclim_dir = save_dict + 'icclim_ig45/icclim_supernrm_agg_daily_time_series_'+ sea + '_' +var1 + '*.nc'
        ccam_dir = save_dict +'ccam_acs/ccam_acs_supernrm_agg_daily_time_series_'+ sea+ '_'+var + '*.nc'
        cmip6_dir = save_dict + 'cmip6/cmip6_supernrm_agg_daily_time_series_'+ sea+ '_'+var + '*.nc'

        agcd[var] = xr.open_dataarray(glob.glob(agcd_dir)[0])
        agcd[var] = agcd[var].sel(time=~((agcd[var].time.dt.month == 2) & (agcd[var].time.dt.day == 29)))
        
        # barpa
        barpa[var] = xr.open_mfdataset(barpa_dir,
                                        combine='nested' ,
                                      concat_dim ='ensemble', preprocess = preprocess).load().sortby('ensemble', 
                                                         ascending=True).__xarray_dataarray_variable__.load()
        # icclim
        icclim[var] = xr.open_mfdataset(icclim_dir,
                                        combine='nested',
                                      concat_dim ='ensemble').sortby('ensemble', ascending=True).__xarray_dataarray_variable__.load()
        icclim[var]['var'] = [var]
        

        # ccam
        ccam[var] = xr.open_mfdataset(ccam_dir,
                                        combine='nested',
                                      concat_dim ='ensemble', preprocess = preprocess).sortby('ensemble', ascending=True).__xarray_dataarray_variable__.load()
        

        # cmip6
        cmip6[var] = xr.open_mfdataset(cmip6_dir,
                                        combine='nested' ,
                                      concat_dim ='ensemble', preprocess = preprocess).sortby('ensemble', ascending=True).__xarray_dataarray_variable__.load()
        
        
        
        # barpa name
        barpa_name = ['BARPA-' + str(x) for x in barpa[var].ensemble.values]
        barpa[var]['ensemble'] = barpa_name

        # ccam name
        ccam_name = ['CCAM-ACS-' + str(x) for x in ccam[var].ensemble.values]
        ccam[var]['ensemble'] = ccam_name
        
        # icclim_name 
        icclim_name = ['QldFCP-2-' + str(x) for x in icclim[var].ensemble.values]
        icclim[var]['ensemble'] = icclim_name
        
        # cmip6 name
        cmip6_name = ['CMIP6-' + str(x) for x in cmip6[var].ensemble.values]
        cmip6[var]['ensemble'] = cmip6_name
        
            
        # gcm[var] = cmip6[var]
        # rcm
        # rcm[var] = xr.concat([barpa[var], icclim[var], ccam[var]], dim = 'ensemble')
    return agcd, barpa,icclim,ccam, cmip6
    
    # return agcd, barpa, ccam, cmip6
```

```python

def get_percentile(ds, obs, threshold):
    """
    to get the time series above threshold
    e.g. when threshold=95, to get the time series above 95th percentile
    Parameters:
    -------
        ds: xarray
            time series of models for each season
        obs: xarray
            time series of models for each season
        threshold: int
            percentile

    Returns
    -------
        new_ds: xarray
            model time series above threshold
    """
    # select the time series exceeding threshold
    p_ds = np.percentile(obs, threshold,-1)[...,np.newaxis]
    
    nt = len(ds[0,0,0,...])
    # broadcast the new time series to compare with the original time series
    new_pds = np.tile(p_ds,[1,1,1,nt])

    # broadcast the new time series to compare with the original time series
    new_ds = np.where( ds>new_pds, ds, np.nan)
    
    return new_ds
```

```python
def calculate_bins_for_all_models(barpa, cmip6, icclim, ccam, agcd, var_name_list, sea,w, threshold = 0):
    """
    calculate_bins_for_all_models
    get the bin edges: including the left edge of the first bin and the right edge of the last bin
    Parameters:
    -------
        agcd: xarray
            daily time series of AGCD for all locations, for each season
        barpa: xarray
            daily time series of BARPA for all locations, all ensembles, and for each season
        ccam: xarray
            daily time series of CCAM for all locations, all ensembles, and for  each season
        icclim: xarray
            daily time series of DES-CCAM for all locations, all ensembles, and for  each season
        cmip6: xarray
            daily time series of CMIP6 for all locations, all ensembles, and for each season
        !!! Narclim: xarray
            daily time series of Narclim for all locations, all ensembles, and for each season
        var_name_list: list
            variables used in this study. ['pr', 'tasmin', 'tasmax']
        sea: string
    Returns
    -------
        agcd: xarray
            daily time series of AGCD for all locations, for each season
        barpa: xarray
            daily time series of BARPA for all locations, all ensembles, and for each season
        ccam: xarray
            daily time series of CCAM for all locations, all ensembles, and for  each season
        icclim: xarray
            daily time series of DES-CCAM for all locations, all ensembles, and for  each season
        cmip6: xarray
            daily time series of CMIP6 for all locations, all ensembles, and for each season
    """
        
    barpa_time = {}
    agcd_time = {}
    cmip6_time = {}
    ccam_time = {}
    icclim_time = {}
    
    
    barpa_time_threshold = {}
    ccam_time_threshold = {}
    cmip_time_threshold = {}
    icclim_time_threshold = {}
    agcd_time_threshold = {}
    bins = {}
    
    for nvar, var_name in enumerate(var_name_list):
        ## remove the mean
        barpa_time[var_name] = barpa[var_name] - barpa[var_name].mean('time')
        agcd_time[var_name] = agcd[var_name] - agcd[var_name].mean('time')
        cmip6_time[var_name] = cmip6[var_name] - cmip6[var_name].mean('time')
        ccam_time[var_name]  = ccam[var_name] - ccam[var_name].mean('time')
        icclim_time[var_name]  = icclim[var_name] - icclim[var_name].mean('time')
        if threshold >0:
            
            barpa_time_threshold[var_name] = get_percentile(barpa_time[var_name], agcd_time[var_name], threshold)
            
            ccam_time_threshold[var_name] = get_percentile(ccam_time[var_name], agcd_time[var_name],threshold)
            cmip_time_threshold[var_name] = get_percentile(cmip6_time[var_name], agcd_time[var_name],threshold)
            icclim_time_threshold[var_name] = get_percentile(icclim_time[var_name], agcd_time[var_name],threshold)
            agcd_time_threshold[var_name]  = get_percentile(agcd_time[var_name], agcd_time[var_name],threshold)
            
           
            bins[var_name]= np.arange(min(np.nanmin(barpa_time_threshold[var_name]),np.nanmin(agcd_time_threshold[var_name]), 
                                      np.nanmin(cmip_time_threshold[var_name]), np.nanmin(ccam_time_threshold[var_name]),
                                      np.nanmin(icclim_time_threshold[var_name])),
                     max(np.nanmax(barpa_time_threshold[var_name]), np.nanmax(agcd_time_threshold[var_name]), 
                         np.nanmax(cmip_time_threshold[var_name]), np.nanmax(ccam_time_threshold[var_name]),
                         np.nanmax(icclim_time_threshold[var_name])) +w, w)   
        ### Full distribution    
        else:
            
            
            bins[var_name]= np.arange(min(np.nanmin(barpa_time[var_name]),np.nanmin(agcd_time[var_name]), 
                                      np.nanmin(cmip6_time[var_name]), np.nanmin(ccam_time[var_name]),
                                      np.nanmin(icclim_time[var_name])),
                     max(np.nanmax(barpa_time[var_name]), np.nanmax(agcd_time[var_name]), 
                         np.nanmax(cmip6_time[var_name]), np.nanmax(ccam_time[var_name]),
                         np.nanmax(icclim_time[var_name])) + w, w)
       
    return bins
```

```python
##### benchmark
importlib.reload(pss)

## pss values for CMIP6 for all ensembles -> full distribution
pss_cmip6_value_all_ensemble = {}

for ns, sea in enumerate(seasons):
    agcd = {}
    cmip6 = {}
    cmip6_pss_loc_var = {}
    for nvar, var_name in enumerate(var_name_list): 
        ## remove the mean
        agcd[var_name] = agcd_time[sea][var_name] - agcd_time[sea][var_name].mean('time')
        cmip6[var_name] = cmip6_time[sea][var_name] - cmip6_time[sea][var_name].mean('time')        
        
        
        nenms = cmip6_time[sea][var_name].ensemble.size
        
        # create the array
        cmip6_pss_with_all_data = np.zeros([nenms, 4])*np.nan
        
        for i,location in enumerate(location_list):
            ## calculate pss value for each var each season each locations
            pss_values_cmip6  = pss.perkins_scores_each_ensemble(cmip6[var_name].sel(location=[location]).squeeze(),
                                                                 agcd[var_name].sel(location=[location]).squeeze(),
                                                                 bins_all_models[sea][var_name]).values
            ## calculate benchmarks
            pss_values_benchmarking[ns, nvar, i]  = np.mean(pss_values_cmip6)
            ## return PSS array for every ensemble to plot the GCM heatmap
            cmip6_pss_with_all_data[:, i] = pss_values_cmip6
        # for all variables    
        cmip6_pss_loc_var[var_name] =  xr.DataArray(cmip6_pss_with_all_data,
                                             dims = ['ensemble','location'],
                                            coords = {'ensemble': cmip6_time[sea][var_name].ensemble, 'location': locations})
     # for all seasons     
    pss_cmip6_value_all_ensemble[sea] =  cmip6_pss_loc_var 

pss_values_benchmarking_array = xr.DataArray(pss_values_benchmarking,
                                             dims = ['season', 'var','location'],
                                            coords = {'season': seasons, 'var': var_name_list, 'location': locations })
```

### Calculate the PSS for RCMs

```python

def pss_one_model(one_model_time,  agcd_time, var_name_list, sea, bins, threshold):
    """
    pss_one_model: calculate PSS values for all ensembles of RCM models
    
    Parameters:
    -------
        one_model_time: xarray
            daily time series of one RCM model for one season
        agcd_time: xarray
            daily time series of AGCD model for one season
        bins: array/list
            bins edge caculated above
        cmip6: xarray
            daily time series of CMIP6 for all locations, all ensembles, and for each season
        threshold: float
            if threshold = 0: full distribution
            
    Returns
    -------
        pss_array: xarray
            PSS value for each season
    """   
    one_model = {}
    agcd = {}
    cmip6 ={}
    pss_array = {}
    for nvar, var_name in enumerate(var_name_list):
        rcm_ensemble_size = one_model_time[var_name].ensemble.size
        pss_values_rcm = np.zeros([4,rcm_ensemble_size])
        agcd[var_name] = agcd_time[var_name].sel(location = location_list) - agcd_time[var_name].mean('time').sel(location = location_list)
        one_model[var_name] = one_model_time[var_name].sel(location = location_list) - one_model_time[var_name].mean('time').sel(location = location_list)
        
            
        ensemble_num = one_model[var_name].ensemble.size
        locations = one_model[var_name].location.values
        for i,location in enumerate(location_list): 
        
            pss_values = pss.perkins_scores_each_ensemble(one_model[var_name].sel(location=[location]).squeeze(), 
                                                          agcd[var_name].sel(location=[location]).squeeze(), bins[var_name], threshold)
            
            pss_values_rcm[i,...] = pss_values

        pss_values_rcm_array = xr.DataArray(pss_values_rcm, dims = ['location', 'ensemble'], coords = {'location': locations, 'ensemble': one_model_time[var_name].ensemble})
        pss_array[var_name]  = pss_values_rcm_array
    return pss_array
```

```python
# for barpa
pss_barpa_var = {}
for ns, sea in enumerate(seasons):
    pss_barpa_var[sea] = pss_one_model(barpa_time[sea],  agcd_time[sea], var_name_list,sea, bins_all_models[sea],0)
# for ccam
pss_ccam_var = {}
for ns, sea in enumerate(seasons):
    pss_ccam_var[sea] = pss_one_model(ccam_time[sea],  agcd_time[sea],  var_name_list,sea, bins_all_models[sea],0)
# for des-ccam
pss_icclim_var = {}
for ns, sea in enumerate(seasons):
    pss_icclim_var[sea] = pss_one_model(icclim_time[sea],  agcd_time[sea], var_name_list,sea, bins_all_models[sea],0)

# pss_narclim_var = {}
# for ns, sea in enumerate(seasons):
#     pss_narclim_var[sea] = pss_one_model(narclim_time[sea],  agcd_time[sea], var_name_list,sea, bins_all_models[sea],0)
```

```python
 # , 'annual_mean'
location_dict = {"Eastern Australia":"EA",
                'Southern Australia':'SA',
                'Rangelands': 'R',
                'Northern Australia': 'NA'}
location_abbre = []
for i in pss_barpa_var['djf']['pr'].location.values:
    location_abbre.append(location_dict[i])
season = ['djf', 'jja']
opt_xb = [True, True]
opt_yb = [True, False]
opt_colorbar = [True, False]
variables = ['pr', 'tasmax', 'tasmin']
def set_ax(var, variable, opt_xb, opt_yb,xlabel,ax,AX, fig, colorbar =True):
    """
    set_ax
    plot the color and set axis
    
    Parameters:
    -------
        var: xarray
            variables to plot
        variable: list
            [pr, tasmax, tasmin]
        opt_xb: whether label the x-axis
        opt_yb: whether label the y-axis
        ylabel: y-axis's label to mark the season
        AX: (AX_left, AX_width)
            where the colorbar starts and the width of colorbar
        colorbar: whether plot the colorbar
    """      
    vmin, vmax = 0.2, 1
    inter = 0.1
    ticks = np.arange(vmin, vmax +inter,inter)
    cmap = 'RdYlBu'   
    c = plt.imshow(var, cmap = cmap, 
                   vmin = vmin, vmax = vmax, 
                   origin ='lower',
                  aspect = 0.7)
    

    ny = var.ensemble
    nx = location_abbre
    
    if opt_yb:
        ft = 5
        ax.set_yticks(np.arange(ny.size))
        ax.set_yticklabels(ny.values, fontsize = ft)
        
    else:
        ax.set_yticks([])
       
   
    if opt_xb:
        ft = 5
        ax.set_xticks(np.arange(len(ny)))
        ax.set_xticklabels(nx, fontsize = ft)
        ax.set_xlabel(xlabel, fontsize = ft+1)
    else:
        ax.set_xticks([])
    # Set ticks on both sides of axes on
    
    ax.tick_params(axis="y", bottom=True, top=False, labelbottom=True, labeltop=False)
    ax.tick_params(axis="x", left=True, right=False, labelleft=True, labelright=False)
    # Rotate and align bottom ticklabels
    plt.setp([tick.label1 for tick in ax.yaxis.get_major_ticks()], rotation=45,
             ha="right", va="center", rotation_mode="anchor")
    plt.setp([tick.label1 for tick in ax.xaxis.get_major_ticks()], rotation=45,
             ha="right", va="center", rotation_mode="anchor")
    # Rotate and align top ticklabels
    if colorbar == True:
        # cb = plt.colorbar(c, extend = 'both', ticks=ticks, pad = 0.02,location ='bottom')
        # cb.ax.tick_params(labelsize=5) 
        ax_l, ax_width = AX
        cax = fig.add_axes([min(ax_l), 0.01, ax_width[0] *2, 0.02])
        cb = plt.colorbar(c,cax=cax, extend = 'min', ticks=ticks,pad = 0.0, orientation = 'horizontal' )
        cb.ax.tick_params(labelsize=5) 
### add the narcli
def agg_pss_benchmarking_plot(rcm_tuple, benchmark_array, variables, figsize):
    """
    agg_pss_benchmarking_plot
    final plot
    
    Parameters:
    -------
        rcm_tuple: tuple
            RCMs to plot (BARPA, CCAM-ACS,.....)
        benchmark_array: xarray
            benchmar array
        variables: list
            [pr, tasmax, tasmin]
        figsize: the size of the figure e.g. (13,14)
    """      

    ###!!!!!!!!!!!!!! addd narclim here
    # pss_barpa_var,ccam_pss_season, icclim_pss_season, narclim_pss_season= rcm_tuple
    pss_barpa_var,ccam_pss_season, icclim_pss_season= rcm_tuple
    fig =plt.figure(figsize=figsize, dpi = 300)
    for nv, var in enumerate(variables):
        
        AX_l, AX_width = [],[]
        
        for ns, seas in enumerate(season):
            # select the seasons
            barpa_pss_season = pss_barpa_var[seas][var_name].transpose( 'ensemble','location')
            ccam_pss_season = pss_ccam_var[seas][var_name].transpose( 'ensemble','location')
            icclim_pss_season=pss_icclim_var[seas][var_name].transpose( 'ensemble','location')
            
            ###!!!!!!!!!!!!!! addd narclim here
            #  narclim_pss_season=pss_narclim_var[seas][var_name].transpose( 'ensemble','location')
            benchmark_ = benchmark_array.sel(season = [seas], var = [var_name]).squeeze()

            #### !!!!!!!!!!!!!! addd narclim here ->
            #  rcm_values_season = xr.concat([barpa_pss_season, icclim_pss_season,ccam_pss_season,,narclim_pss_season,benchmark_], dim = 'ensemble')
            
            rcm_values_season = xr.concat([barpa_pss_season, icclim_pss_season,ccam_pss_season,benchmark_], dim = 'ensemble')
            ax=plt.subplot(1,8,nv*3+1+ns)
            
            left, bottom, width, top = ax.get_position().bounds
            AX_l.append(left)
            AX_width.append(width)
            
            x = 3.5

            ### x1, x2, x3, x4, the location of the subtitle of RCM's name and 'Benchmark'
            y1 = 2
            y2 = 11
            y3 = 18.5
            y4 = 23

            ### x_line1, x_line2 mark the location of boundary line of each RCM and Benchmark column
            x_line1 = 6.5
            x_line2 = 16.5
            ft = 6
            
            if (ns == 1) & (nv == 2):
               
                ax.text(x, y1,'BARPA-ACS', fontsize = ft, rotation = 270)
                ax.text(x, y2, 'QldFCP-2', fontsize = ft, rotation = 270)
                ax.text(x, y3,'CCAM-ACS', fontsize = ft, rotation = 270)
                # ax.text(x, y4,'NARCLIM', fontsize = ft, rotation = 270)
        
            xlabel = seas.upper()
            alpha = string.ascii_lowercase[nv]
            plt.axhline(x_line1, c = 'k')
            plt.axhline(x_line2, c = 'k')
            plt.axhline(23.5, c = 'k')
            # plt.axhline(narclim_x, c = 'k')
           
            ### Title: variable's name
            ### a) Pr (%)...
            if ns ==0:
                colorbar = False
                y = rcm_values_season.ensemble.size
                if var  == 'pr':
                    
                    ax.text(-0.7, y, alpha + ') ' + 'pr (%)', fontsize = 7,  weight = 'bold', ha="center", va = 'bottom')
                else:
                    ax.text(-0.7, y, alpha + ') ' + var, fontsize = 7,  weight = 'bold', ha="center", va = 'bottom')
            else:
                colorbar = True
                

            set_ax(rcm_values_season,var, opt_xb[ns], opt_yb[ns],xlabel, ax,  (AX_l, AX_width),fig, colorbar =colorbar)
            nx = len(location_abbre)
            ny = rcm_values_season.ensemble.size    
            
            ### values of each RCM 
            for i in range(ny):
              for j in range(nx):
                  fill_values = "{:.2f}".format(rcm_values_season[i,j].values) 
                  if rcm_values_season[i,j].values < benchmark_[j].values:
                       ## if RCM ensembles don't meet the benchmark
                      ##The text will be of different pattern 
                            ## with bbox=dict(facecolor='lavender', edgecolor='blueviolet',boxstyle='round')
                            ## to identify the different style
                      ax.text(j, i,fill_values, ha="center", va = 'center',fontsize = 4.2, 
                      color = 'k', weight = 'bold', bbox=dict(facecolor='lavender', edgecolor='blueviolet',boxstyle='round'))
                  else: ## if RCM ensembles meet the benchmark, the text will be of plain pattern 
                      ax.text(j, i,fill_values, ha="center", va = 'center', fontsize = 5)
            plt.subplots_adjust(wspace = 0.01, hspace = None, bottom = None, top = None, left = None, right = None)
            
    
        ### save the fig
        
    plt.savefig('/g/data/tp28/dev/xj2271/acs/ACS_Figure/pss_superngm_updated_1985_2014_rcm_update', bbox_inches = 'tight')
    
```
